```python
import random #importing the random library for randomly generating numbers to be inserted
class Node: #class for a node
    def __init__(self,key):
        self.left=None
        self.right=None
        self.data=key


def postorder(root): #method for printing out the postorder traversal for the BST, takes the root as input
    if root is None:
        return
    print(root.data,end=" ")
    if root.left is not None:  
        postorder(root.left)    
    if root.right is not None:
        postorder(root.right)
        
        
        
def inorder(root): #method for printing out the inorder traversal for the BST, takes the root as input
    if root is None:
        return
    if root.left is not None:  
        inorder(root.left)
    print(root.data,end=" ")
    if root.right is not None:
        inorder(root.right)
        
        
def preorder(root): #method for printing out the preorder traversal for the BST, takes the root as input
    if root is None:

        return
    if root.left is not None:  
        preorder(root.left)
    if root.right is not None:
        preorder(root.right)
    print(root.data,end=" ")


def search(root,key): #method for searching a node(key) in the BST
    if root is None:
        return None
    else:
        if root.data==key:
#returning root so that it can be used in function like left_child etc            
            return root                 
        elif root.data <key:            
            return search(root.right,key)
        else:
            return search(root.left,key)
        


def insert(root,key): #method for inserting a node(key) to the BST
    if root is None:
        return Node(key)
    else:
        if root.data< key:
            root.right = insert(root.right,key)
        else:
            root.left = insert(root.left,key)
    return root

def find_min(root):
    if root is None or root.left is None:
        return root
    return find_min(root.left)

def find_max(root):
    if root is None or root.right is None:
        return root
    return find_max(root.right)

def left_child(root,key):
    node=search(root,key)
    if node is None:
        print(str(key)+" is not present in the tree...")
    elif node.left is None:
        print(str(key)+" has no left child...")
    else:
        print(str(key)+" has left child as: "+str(node.left.data))

def right_child(root,key):
    node =search(root,key)
    if node is None:
        print(str(key)+" is not present in the tree...")
    elif node.right is None:
        print(str(key)+" has no right child...")
    else:
        print(str(key)+" has right child as: "+str(node.right.data))

def count_total(root):
    if root is None:
        return 0
    return 1+count_total(root.left)+count_total(root.right)

def count_leaf(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    else:
        return count_leaf(root.left)+count_leaf(root.right)
    
def count_full(root):
    if root is None:
        return 0
    if root.left is not None and root.right is not None:
        return 1+count_full(root.left)+count_full(root.right)
    else:
        return 0


def print_ancestors(root,key):
    if root is None:
        return False
    if root.data==key:
        print(f"Ancestors of {key} are: ",end=" ")
        print(str(root.data)+" -> ",end=" ")
        return True
    if (print_ancestors(root.left,key) or print_ancestors(root.right,key)):
        print(str(root.data)+" -> ",end=" ")
        return True
    return False

def print_descendants(root,key):
    node = search(root,key)
    if node is None:
        print(str(key)+" is not present in tree!!!!")
    else:
        print(f"Descendant nodes of {key} when traversed using in-order: ",end=" ")
        inorder(node)

def deleteNode(root, key):
    if root is None:
        return root
    if key < root.data:
        root.left = deleteNode(root.left, key)
    elif(key > root.data):
        root.right = deleteNode(root.right, key)
    else:
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        temp = find_min(root.right)
        root.dat = temp.data
        root.right = deleteNode(root.right, temp.data)
 
    return root
def delete(root,key):
    node = search(root,key)
    if node is None:
        print("Key not found!!!")
        return
    deleteNode(root,key)    



BST= Node(120)

for i in range(40):
    insert(BST,random.randint(1,300))
print("Total number of nodes in tree: "+str(count_total(BST))+"\n")
print("Total  leaf nodes in tree: "+str(count_leaf(BST))+"\n")
print("Number of full nodes in tree: "+str(count_full(BST))+"\n")
print("Min element is: "+str(find_min(BST).data)+"\n")
print("Max element is: "+str(find_max(BST).data)+"\n")

while(1):
    x= random.randint(1,250)
    if search(BST,x) is not None:
        break
print(f"Randomly selected element {x} in the tree. ")
print()
left_child(BST,x)
print()
right_child(BST,x)
print()
print_ancestors(BST,x)
print("\n")
print_descendants(BST,x)
print()
delete(BST,x)
print()
print("Inorder after deleting "+str(x)+" is: ")
inorder(BST)
```

```python
import random
class Disjoint:
    def __init__(self, vertex):
        self.rank=[]
        self.parent=[]
        for node in range(vertex):
            self.parent.append(node)
            self.rank.append(0)
    
    def find(self,i): #finding the parent of the node
        if self.parent[i]==i:
            return i
        return self.find(self.parent[i])
    
    def union(self,x,y): #method for performing union of x and y
        x_parent = self.find(x)
        y_parent=self.find(y)

        if self.rank[x_parent]<self.rank[y_parent]:
            self.parent[x_parent]=y_parent
        elif self.rank[x_parent]>self.rank[y_parent]:
            self.parent[y_parent]=x_parent
        else:
            self.parent[x_parent]=y_parent
            self.rank[y_parent]+=1

class Graph(Disjoint):
    def __init__(self,vertex):
        self.vertices=vertex
        self.graph=[]
    
    def addEdge(self,u,v,w): #adding an edge to the graph
        self.graph.append([u,v,w])
    
    def display(self,result):
        cost=0
        for item in result:
            print(f'{item[0]} -- > {item[1]} with weight {item[2]}')
            cost+=item[2]
        print(f'Total cost is {cost}')
    
    def Kruskal(self): #performing krushkal's algorithm on the given graph
        result=[]
        self.graph=sorted(self.graph,key=lambda x:x[2])
        Set=Disjoint(self.vertices)
        i=0
        edge=0
        while edge<self.vertices -1:
            u,v,w=self.graph[i]
            x=Set.find(u)
            y=Set.find(v)
            if x!=y:
                if [v,u,w] not in result:
                    result.append([u,v,w])
                    edge=edge+1
                    Set.union(x,y)
            i+=1
        
        self.display(result)

graph = Graph(7)
graph.addEdge(0,2,51)
graph.addEdge(0,1,54)
graph.addEdge(1,5,67)
graph.addEdge(0,3,40)
graph.addEdge(2,4,25)
graph.addEdge(2,3,32)
graph.addEdge(3,5,89)
graph.addEdge(4,5,78)
graph.addEdge(4,6,95)
graph.addEdge(5,6,56)

graph.Kruskal()
```

```python
from math import trunc


class Heap:
    def __init__(self,type,name,size):
        self.heap=[]
        self.type=type
        self.name=name
        self.size=size

    def heapify(self,pos):
        curr = pos
        l_child = 2*pos+1
        r_child = 2*pos +2
        if self.type=="Max":
            if l_child<len(self.heap) and self.heap[curr]<self.heap[l_child]:
                curr=l_child
            if r_child < len(self.heap) and self.heap[curr]<self.heap[r_child]:
                curr=r_child

        if self.type=="Min":
            if l_child<len(self.heap) and self.heap[curr]>self.heap[l_child]:
                curr=l_child
            if r_child < len(self.heap) and self.heap[curr]>self.heap[r_child]:
                curr=r_child

        if curr != pos:
            self.heap[curr],self.heap[pos]=self.heap[pos],self.heap[curr]
            self.heapify(curr)
            
    def shift_up(self,pos):
        par =(pos-1)//2
        if par<0:
            return
        swap=False
        if self.type=="Max":
            if self.heap[par]< self.heap[pos]:
                swap=True
        else:
            if self.heap[par]>self.heap[pos]:
                swap=True
        if swap:
            self.heap[par],self.heap[pos]=self.heap[pos],self.heap[par]
            self.shift_up(par)

    def insert(self,val):
        if self.size <= len(self.heap):
            print("Can't be added... Heap is Full")
            return -1
        self.heap.append(val)
        self.shift_up(len(self.heap)-1)
        return 0

    def extract_max(self):
        if len(self.heap)<=0:
            print("Underflow!!!!")
            return None
        temp=self.heap[0]
        self.heap[0]=self.heap[len(self.heap)-1]
        self.heap.pop()
        self.heapify(0)
        return temp

    def Print(self):
        print(f"\n Heap named: {self.name} ")
        for i in range(0, (len(self.heap) // 2)):
            print(" PARENT : " + str(self.heap[i]),end=" ")
            if 2*i+1 <len(self.heap):
                print(" LEFT CHILD : " + str(self.heap[2 * i+1]),end=" ")
            if 2*i+2 <len(self.heap):
                print( " RIGHT CHILD : " + str(self.heap[2 * i + 2]),end=" ")
            print()

def AddElements(ar):
    print("Enter Elements one by one into the heap: ",end=" ")
    print("Enter X when done",end=" ")
    while True:
        ele=(input())
        if ele=="X" or ele=="x":
            break
        z=ar.insert(int(ele))
        if z== -1:
            print("Press any key to continue...")
            break

def Child_Menu(curr):
    print(f'Name of selected heap: {curr.name}')
    print(f'Type: {curr.type}')
    while True:
        choice=int(input(f"1. Insert    2. Extract {curr.type}  3. Display Heap    4. Return to main Menu:    "))
        if choice==1:
            AddElements(curr)
        elif choice ==2:
            x= curr.extract_max()
            if x is not None:
                print("Extracted element is "+str(x))
        elif choice==3:
            curr.Print()
        elif choice==4:
            return
        else:
            print("Please Enter correct input... Press any key to try again!!")
            print("\n")
        print("\n")



heaps=[]
while True:

    print("1. Create Min Heap")
    print("2. Create Max Heap")
    print("3. Select an already created Heap")
    choice=int(input("4. Exit (Will delete all the created Heaps):    "))
    if choice==1 or choice==2:
        name=input("Type a name so that it can be accessed later on too:    ")
        size=int(input(f"Enter the size of heap named {name}:    "))
        if choice==1:
            heaps.append(Heap("Min",name,size))
        else:
            heaps.append(Heap("Max",name,size))
        AddElements(heaps[-1])
        choose=len(heaps)-1
    elif choice ==3:
        count=0
        print("Enter the index corresponding to one")
        for i in heaps:
            print(f'{count} {i.name}')
            count+=1
        choose =int(input())
        if choose>len(heaps):
            continue
    elif choice==4:
        break
    else:
        print("Please Enter the correct number...")
        print("Press any key...")
        for i in range(100):
            print("\n")
        continue
    Child_Menu(heaps[choose])
    print("\n")

```

```python
from itertools import permutations
import sys
class Graph:
    def __init__(self,size): #constructor for initialising the graph
        self.size=size
        self.graph=[]
    
    def addVertex(self,weights): #adding a vertex to the graph
        self.graph.append(weights)
    def TSP(self,src): #performing the travelling salesman from the source
        min_cost = sys.maxsize
        min_path=[]
       
        neighbors =[i for i in range(self.size) if i !=src]
        
        permute=permutations(neighbors)
        for i in permute:
            path=[]
            path.append(src)
            curr=0
            k=src
            for vertex in i:
                curr = curr+self.graph[k][vertex]
                path.append(vertex)
                k=vertex
            curr=curr+self.graph[k][src]
            path.append(src)
            min_cost = min_cost if min_cost<curr else curr
            min_path=path if curr==min_cost else min_path
        return min_cost,min_path

graph= Graph(4)
graph.addVertex([0, 150, 125, 210])
graph.addVertex([55, 0, 235, 210])
graph.addVertex([66, 123, 0, 121])
graph.addVertex([85, 82, 79, 0])
cost,path=graph.TSP(0)
print(f'Least cost possible is: {cost}')
print("Path followed is: ",end=" ")
print(str(path[0]),end=" ")
for i in range(len(path)-1):
    print(f' -> {path[i+1]}',end=" ")


```
